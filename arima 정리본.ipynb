```python
def tsplot(y,lags=None,figsize=(10,8),style='bmh',title=''):
    if not isinstance(y,pd.Series):
        y=pd.Series(y)
    with plt.style.context(style):
        fig=plt.figure(figsize=figsize)
        layout=(3,2)
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))
        qq_ax = plt.subplot2grid(layout, (2, 0))
        pp_ax = plt.subplot2grid(layout, (2, 1))
        
        y.plot(ax=ts_ax)
        ts_ax.set_title(title)
        smt.graphics.plot_acf(y,lags=lags,ax=acf_ax,alpha=0.5)
        smt.graphics.plot_pacf(y,lags=lags,ax=pacf_ax,alpha=0.5)
        sm.qqplot(y,line='s',ax=qq_ax)
        qq_ax.set_title('QQ plot')
        scs.probplot(y,sparams=(y.mean(),y.std()),plot=pp_ax)
        
        plt.tight_layout()
    
    return

```

# p,q는 어떻게 구할것인가?



p와 q는 0이 될수 있고 그 이상을 가지고 갈수 있는데 
만약 p와 q가 0보다 크다면 이것은 미래가 더 불확실 하다는 의미를 말한다

규칙1.
>Rule 6: If the PACF of the differenced series displays a sharp cutoff and/or the lag-1 autocorrelation is positive--i.e., if the series appears slightly "underdifferenced"--then consider adding an AR term to the model. The lag at which the PACF cuts off is the indicated number of AR terms.

>>pacf에서 cuf_off를 보이거나 (lag-1) autocorrelation 이 양의 가진다면(다시 말해서
시리즈가 약간 "차이가 덜한것처럼" 보인다면 그 lag 값은 ar위치에 한다


자기 상관 함수(ACF)는 PACF가 AR항에 대해 수행하는 MA항에 동일한 역할을 할수 있다.
다시말해서 ACF는 차이 계열에서 나머지 자기 상관을 제거하는데 필요한 MA 항의 수를 알려 줄 수 있다.


규칙2.
> If the ACF of the differenced series displays a sharp cutoff and/or the lag-1 autocorrelation is negative--i.e., if the series appears slightly "overdifferenced"--then consider adding an MA term to the model. The lag at which the ACF cuts off is the indicated number of MA terms.

>> 급격하게 ACF에서 변화하는 부분 또는 LAG-1 에서 자기 상관이 음수 인경우(즉, 급격한 차이를 보여준다면) 모델에 MA 에 값을 추가하자. ACF가 끊어진다는것은 결국 오차가 발생한다는 것이고 그 오차를 설명하는 것은 MA다,



규칙3
>It is possible for an AR term and an MA term to cancel each other's effects, so if a mixed AR-MA model seems to fit the data, also try a model with one fewer AR term and one fewer MA term--particularly if the parameter estimates in the original model require more than 10 iterations to converge.

>>어쩌면 AR과 MR 을 혼합하는것은 서로의 효과를 상쇄 시킬수 있지만 떄로는 아닐수도 있어서 적어도 10회 이상 동안 AR과 MA값에 설정된 값들을 변화를 주면서 계속 찾아봐야 한다



모델은 어떤걸로?
> 시각적 차이가 있겠지만 차분이 하나인것은 평균적인 추세를 가정하고 있으며, 상대적으로 예측값의 정확도에 대해서 낙관적인 전망을 봐라 볼수 있지만 두 차수를 가지고 있는것은 아무래도 두가지 측면에 대해서 고려하다보니 상대적으로 정확도가 낮게 평가 될 수 밖에 없다. 그래서 결론적으로 일반적인 규칙으로 차분 차수가 낮은 모델을 선택하는 것이 올바르다고 할수 있다


규칙4
> If there is a unit root in the AR part of the model--i.e., if the sum of the AR coefficients is almost exactly 1--you should reduce the number of AR terms by one and increase the order of differencing by one.

>>만약 ar에 대한 가중치 합이 1이 이라면 lag 값을 늘리고 또다른 가중치를 줄여본다


규칙5
>If there is a unit root in the MA part of the model--i.e., if the sum of the MA coefficients is almost exactly 1--you should reduce the number of MA terms by one and reduce the order of differencing by one.

>> 만약 ma에 대한 가중치 합이 거의 1과 가깝다면 ma의 lag값을 늘리고 또다른 가중치를 줄여본다


규칙6
> If the long-term forecasts appear erratic or unstable, there may be a unit root in the AR or MA coefficients 

>> 만약 장기 예측 값들이 불규칙 해 보이거나 불안정 해보이면 ar 또는 ma 계수에 단위근이 있을 수있다
